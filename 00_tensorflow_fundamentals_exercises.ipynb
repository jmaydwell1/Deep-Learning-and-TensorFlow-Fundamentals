```python
 ones = tf.ones(shape=(5, 300))
```

```python
zeros = tf.zeros(shape=(5, 300))
```

```python
ones * zeros
```

```python
tf.multiply(ones, zeros)
```

```python
zeros_reshaped = tf.reshape(zeros, shape=(300, 5))
```

```python
ones @ zeros_reshaped
```

```python
tf.transpose(ones)
```

```python
tf.matmul(tf.transpose(ones), zeros)
```

```python
tf.tensordot(tf.transpose(ones), zeros, axes=1)
```

```python
random_tensor = tf.random.Generator.from_seed(42)
random_tensor = random_tensor.uniform((224, 224, 3), minval=0, maxval=1)
random_tensor 
```

```python
min = tf.reduce_min(random_tensor, axis=1)
max = tf.reduce_max(random_tensor, axis=1)
min, max
```
